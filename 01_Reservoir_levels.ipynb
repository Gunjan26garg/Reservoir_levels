```python
chennai['Date']=pd.to_datetime(chennai['Date'],format='%d-%m-%Y')
```

```python
chennai[['POONDI','CHOLAVARAM','REDHILLS','CHEMBARAMBAKKAM']] = chennai[['POONDI','CHOLAVARAM','REDHILLS','CHEMBARAMBAKKAM']].apply(pd.to_numeric)
```

```python
chennai
```

```python
chennai.info()
```

```python
chennai['Year']=chennai['Date'].apply(lambda x:x.year)
chennai['MONTH']=chennai['Date'].apply(lambda x:x.month)
```

```python
chennai['Total']= chennai[['POONDI','CHOLAVARAM','REDHILLS','CHEMBARAMBAKKAM']].sum(axis=1)
```

```python
chennai
```

```python
YearWise = chennai.groupby('Year',as_index=False)
```

```python
YearWise.head()
```

```python
YearWise.mean()
```

```python
sns.barplot(x='Year',y='Total',data=YearWise.mean())
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x='Year',y='Total',data=YearWise.mean())
```
